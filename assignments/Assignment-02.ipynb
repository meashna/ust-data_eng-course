Because **local variables only exist inside the function** where they are created. Once the function finishes, they are gone. This is called **scope** — a variable's 'zone' where it can be used.

```python
def my_function():
    local_var = 10   # this only exists inside the function
    print(local_var) # works fine here

my_function()

# Trying to use it outside will cause an error:
# print(local_var)  # NameError: name 'local_var' is not defined
```

## 2. map, filter, reduce — When to Use Each?

```python
# MAP - apply same operation to every element
numbers = [1, 2, 3]
squared = list(map(lambda x: x**2, numbers))
print("map (square each):", squared)  # [1, 4, 9]
```

```python
# FILTER - keep only elements that match a condition
numbers = [1, 2, 3]
odds = list(filter(lambda x: x % 2 == 1, numbers))
print("filter (odd only):", odds)  # [1, 3]
```

```python
# REDUCE - combine all elements into a single result
from functools import reduce
numbers = [1, 2, 3]
total = reduce(lambda x, y: x + y, numbers)
print("reduce (sum):", total)  # 6
```

## 3. Simple Calculator

```python
def add(a, b):  return a + b
def sub(a, b):  return a - b
def mul(a, b):  return a * b
def div(a, b):
    if b == 0:
        return "Error: Can't divide by zero!"
    return a // b

print("Simple Calculator")

while True:
    a = int(input("Number 1: "))
    b = int(input("Number 2: "))
    print("1. Add  2. Subtract  3. Multiply  4. Divide  5. Exit")
    choice = int(input("Your choice: "))

    if   choice == 1: print("Result:", add(a, b))
    elif choice == 2: print("Result:", sub(a, b))
    elif choice == 3: print("Result:", mul(a, b))
    elif choice == 4: print("Result:", div(a, b))
    elif choice == 5:
        print("Bye!")
        break
    else:
        print("Invalid choice. Try again!")
```
