```python
import os

print("Current location:", os.getcwd())
```

```python
os.mkdir("my_folder")  # creates a new folder called 'my_folder'
print("Folder created!")
```

```python
os.chdir("my_folder")  # move into the folder
print("Now inside:", os.getcwd())
```

```python
# 'w' = write mode (creates the file if it doesn't exist)
with open("my_file.txt", "w") as f:
    f.write("Hello! This is my first file.\n")
    f.write("Learning Python file handling!")

print("File written!")
```

## Step 5: Read the File

```python
# 'r' = read mode
with open("my_file.txt", "r") as f:
    content = f.read()

print("File content:")
print(content)
```

```python
os.remove("my_file.txt")  # deletes the file
print("File deleted!")
```

## Step 7: Go Back and Delete the Directory

```python
import shutil

os.chdir("..")  # go back one level (out of my_folder)

shutil.rmtree("my_folder")  # delete the entire folder
print("Folder deleted!")
print("Back to:", os.getcwd())
```
