```python
class MyNumbers:
    def __iter__(self):
        self.num = 1
        return self

    def __next__(self):
        if self.num <= 5:
            x = self.num
            self.num += 1
            return x
        else:
            raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
    print(x)
```

```python
Generator -> Special type of iterator , using function and yield keyword
```

```python
def my_gen():
    for i in range(1,6):
        yield i
gen = my_gen()
for value in gen:
    print(value)
```

```python
def test_duplicate_account_number(self):
    bs = Bank()
    bs.add_user("alice", 1234567890, 100)
    with self.assertRaises(ValueError):
        bs.add_user("bob", 1234567890, 50)   # same acc no
 
def test_invalid_account_number_type(self):
    bs = Bank()
    with self.assertRaises(TypeError):
        bs.add_user("alice", "1234567890", 100)
 
def test_negative_account_number(self):
    bs = Bank()
    with self.assertRaises(ValueError):
        bs.add_user("alice", -12345, 100)
 
def test_wrong_account_number_on_deposit(self):
    bs = Bank()
    bs.add_user("alice", 1111111111, 100)
    with self.assertRaises(ValueError):
        bs.deposit("alice", 2222222222, 10)
```

```python
import math
 
class Bank:
    _used_account_numbers = set()   # class-level: to ensure uniqueness
 
    def __init__(self):
        self._users = {}  # username -> {"acc_no": int, "balance": float}
 
    def add_user(self, username, account_number, initial_balance=0):
        # ---- username checks ----
        if not isinstance(username, str):
            raise TypeError("Username must be a string")
        if not username.strip():
            raise ValueError("Username cannot be empty")
        if username.isdigit():
            raise ValueError("Username cannot be only numbers")
        if len(username) > 15:   # username length constraint
            raise ValueError("Username is too long")
 
        # ---- account number checks ----
        if not isinstance(account_number, int):
            raise TypeError("Account number must be an integer")
        if account_number <= 0:
            raise ValueError("Account number must be positive")
        if account_number in Bank._used_account_numbers:
            raise ValueError("Account number already exists")
 
        # ---- balance checks ----
        if not isinstance(initial_balance, (int, float)):
            raise TypeError("Initial balance must be a number")
        if math.isnan(initial_balance) or math.isinf(initial_balance):
            raise ValueError("Initial balance must be finite")
        if initial_balance < 0:
            raise ValueError("Initial balance cannot be negative")
 
        # ---- duplicate user ----
        if username in self._users:
            raise ValueError("User already exists")
 
        # ---- create user ----
        self._users[username] = {
            "acc_no": account_number,
            "balance": float(initial_balance)  # private balance
        }
        Bank._used_account_numbers.add(account_number)
 
    def deposit(self, username, account_number, amount):
        self._validate_user(username, account_number)
        self._validate_amount(amount, "Deposit amount")
 
        self._users[username]["balance"] += float(amount)
 
    def withdraw(self, username, account_number, amount):
        self._validate_user(username, account_number)
        self._validate_amount(amount, "Withdraw amount")
 
        if amount > self._users[username]["balance"]:
            raise ValueError("Insufficient balance")
 
        self._users[username]["balance"] -= float(amount)
 
    def get_balance(self, username):
        if username not in self._users:
            raise KeyError("User does not exist")
        return self._users[username]["balance"]
 
    # -------- Helpers --------
 
    def _validate_user(self, username, account_number):
        if username not in self._users:
            raise KeyError("User does not exist")
 
        if self._users[username]["acc_no"] != account_number:
            raise ValueError("Invalid account number for this user")
 
    def _validate_amount(self, amount, name):
        if not isinstance(amount, (int, float)):
            raise TypeError(f"{name} must be a number")
        if math.isnan(amount) or math.isinf(amount):
            raise ValueError(f"{name} must be finite")
        if amount < 0:
            raise ValueError(f"{name} must be positive")
```
