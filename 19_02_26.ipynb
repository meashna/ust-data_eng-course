```python
# ASSESMENT - FACTORS OF NUMBERS
def hcf(a, b):
    smaller = min(a, b)
    for i in range(smaller, 0, -1): 
        if a % i == 0 and b % i == 0:
            return i

num1,num2=6,8

print(f"HCF of {num1} and {num2} is: {hcf(num1, num2)}")
```

```python
class Sport:
    def __init__(self, model, cost, color):
        self.model = model
        self.cost = cost
        self.color = color

    def getModel(self):
        return self.model

    def getCost(self):
        return self.cost

    def getColor(self):
        return self.color


class Shoes(Sport):
    def __init__(self, model, cost, color, brand):
        super().__init__(model, cost, color)
        self.brand = brand

    def Shoesinfo(self):
        return f"Model: {self.model}, Cost: {self.cost}, Color: {self.color}, Brand: {self.brand}"


sh1 = Shoes("Shoes", "Rs.3699", "Black", "Xyz123")
#print(sh1.getModel())
print(sh1.Shoesinfo())
```
