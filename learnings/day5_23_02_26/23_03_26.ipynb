```python
import pandas as pd
df =pd.read_csv(r"ipl.csv")
df
```

```python
print(df.shape)
print(df.head())
print(df.isnull().sum())
```

```python
wins = df['winner'].value_counts().head(8)


plt.bar(wins.index, wins.values, color='orange')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.title('Most Wins - IPL All Time')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
top_players = df['player_of_match'].value_counts().head(10)


plt.barh(top_players.index[::-1], top_players.values[::-1], color='skyblue')
plt.xlabel('Awards')
plt.title('Top 10 Player of the Match Winners')
plt.tight_layout()

plt.show()
```
