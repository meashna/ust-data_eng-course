```python
import matplotlib.pyplot as plt

categories = ['Housing', 'Transportation', 'Food', 'Entertainment', 'Utilities']
expenses = [1200, 300, 400, 150, 250]  

plt.figure(figsize=(5, 5))

plt.pie(
    expenses,
    labels=categories,
    autopct='%2.2f%%'
)

plt.title("Pie Example")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

quarters = ['Q1', 'Q2', 'Q3', 'Q4']

electronics = [10000, 12000, 11000, 105000]
clothing = [5000, 6000, 7500, 8000]
home_appliances = [3000, 7500, 8200, 9000]

plt.figure(figsize=(10, 6))

plt.stackplot(
    quarters,
    electronics,
    clothing,
    home_appliances,
    labels=['Electronics', 'Clothing', 'Home Appliances']
)

plt.xlabel('Quarters')
plt.ylabel('Sales')
plt.title('Quarterly Sales by Category')

plt.legend(loc='upper left')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Sine
axs[0, 0].plot(x, y1, 'r')
axs[0, 0].set_title('Sine Wave')

# Cosine
axs[0, 1].plot(x, y2, 'b')
axs[0, 1].set_title('Cosine Wave')

# Tangent
axs[1, 0].plot(x, y3, 'g')
axs[1, 0].set_title('Tangent Wave')

# Exponential
axs[1, 1].plot(x, y4, 'm')
axs[1, 1].set_title('Exponential Decay')

plt.tight_layout()
plt.show()
```

```python
# Bar chart: Total revenue by department
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('healthcare_data.csv')

# Group by department and sum bill_amount
revenue_by_dept = df.groupby('department')['bill_amount'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8,5))
revenue_by_dept.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Department')
plt.ylabel('Total Revenue (bill amount)')
plt.title('Total Revenue by Department')
plt.tight_layout()
plt.show()
```
