```python
#Iterator -> obj allows you to traverse all the elements of a collection 
```

```python
class MyNumbers:
    def __iter__(self):
        self.num = 1
        return self

    def __next__(self):
        if self.num <= 5:
            x = self.num
            self.num += 1
            return x
        else:
            raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
    print(x)
```

```python
numbers = [1, 2, 3]
iterator = iter(numbers)
print(next(iterator))  
print(next(iterator))  
print(next(iterator))  
```

```python
Generator -> Special type of iterator , using function and yield keyword
```

```python
def my_gen():
    for i in range(1,6):
        yield i
gen = my_gen()
for value in gen:
    print(value)
```

```python
Iterator vs Generator(comparison)
iterator -> requires class with __iter() and __next__()
Gen -> simple and concise impl
```
